```python
import numpy as np
from scipy.stats import binom
from scipy.integrate import quad
from scipy.stats import norm
```

```python
def likelihood(x):
    return binom.pmf(k=30, n=100, p=x)

x = np.linspace(0, 1, 100)
y = likelihood(x)
evid = quad(likelihood, 0, 1)[0]
y = y / evid
import matplotlib.pyplot as plt
plt.plot(x, y)
x_max_post = x[np.argmax(y)]
print('max likelihood p : ', x_max_post)
plt.vlines(x_max_post, 0, np.max(y), colors='r', linestyles='dashed', label = f'max likelihood : {x_max_post:.3f}')
plt.legend()
plt.show()
```

```python
y2 = likelihood(x) * norm.pdf(x, loc=0.35, scale=0.05)
evid2 = quad(lambda x : likelihood(x) * norm.pdf(x, loc=0.35, scale=0.05), 0, 1)[0]
y2 = y2 / evid2
plt.plot(x, y2)
x_max_post2 = x[np.argmax(y2)]
plt.vlines(x_max_post2, 0, np.max(y2), colors='r', linestyles='dashed', label = f'max a posteriori : {x_max_post2:.3f}')
plt.legend()
plt.show()
print('max a posteriori p : ', x_max_post2)
```
